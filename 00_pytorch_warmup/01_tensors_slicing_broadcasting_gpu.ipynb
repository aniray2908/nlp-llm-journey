```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using: {device}")
```

```python
x = torch.tensor([1.0, 2.0, 3.0], device=device)
y = torch.tensor([4.0, 5.0, 6.0], device=device)
z = x + y
z
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
a = torch.tensor([1.0, 2.0]).to(device)  # GPU
b = torch.tensor([3.0, 4.0])             # CPU

a + b
```

```python
def safe_matmul(a, b):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    a = a.to(device)
    b = b.to(device)
    return torch.matmul(a, b)

a = torch.randn(3, 4)
b = torch.randn(4, 5)
print(safe_matmul(a, b).shape)  # should be (3, 5)
```
